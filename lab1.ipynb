# Моделирование базовой случайной величины мультипликативно-конгруэнтным методом и методом МакЛарена-Марсальи.

```python
import seaborn as sns

from utils.distributions import (
    maclaren_marsaglia,
    multiplicative_congruential,
    uniform_cdf,
)
from utils.tests import (
    chi2_continious,
    ks_test,
)


n = 1000
alpha = 0.05

M = 2 ** 31
a0 = 16387
beta = 16387
K = 48
```

```python
arr1 = multiplicative_congruential(a0, beta, M, n)

chi2_continious(arr1, uniform_cdf)
print()
ks_test(arr1, uniform_cdf)

sns.distplot(arr1)
```

```python
arr2 = multiplicative_congruential(a0, 4*beta+1, M, n)

chi2_continious(arr2, uniform_cdf)
print()
ks_test(arr2, uniform_cdf)

sns.distplot(arr2)
```

```python
arr3 = maclaren_marsaglia(arr1, arr2, K, n)

chi2_continious(arr3, uniform_cdf)
print()
ks_test(arr3, uniform_cdf)

sns.distplot(arr3)
```
